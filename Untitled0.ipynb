```python
# ============================================================
# COA207 – Foundations of AI
# Portfolio: Domain 2 – Route Finding (Search)
# Edinburgh Campus Emergency Route Finder
# ============================================================
# Run each cell in order. No installations required.
# ============================================================


# ── CELL 1: Map ─────────────────────────────────────────────

campus_map = {
    'Main_Gate':       {'Reception': 3, 'Car_Park': 5, 'South_Lodge': 4},
    'Reception':       {'Main_Gate': 3, 'Admin_Block': 4, 'Lecture_Hall_A': 6},
    'Car_Park':        {'Main_Gate': 5, 'Sports_Centre': 7},
    'Admin_Block':     {'Reception': 4, 'Library': 5, 'Lecture_Hall_A': 3},  # ONE-WAY OUT
    'Library':         {'Study_Hub': 4, 'Cafe': 6, 'Lecture_Hall_B': 8},    # no Admin_Block key
    'Study_Hub':       {'Library': 4, 'Lecture_Hall_B': 5, 'Cafe': 3},      # BRIDGE BLOCKED
    'Cafe':            {'Study_Hub': 3, 'Library': 6, 'Science_Block': 5},
    'Lecture_Hall_A':  {'Admin_Block': 3, 'Reception': 6, 'Lecture_Hall_B': 4},
    'Lecture_Hall_B':  {'Lecture_Hall_A': 4, 'Study_Hub': 5,
                        'Science_Block': 6, 'Library': 8},
    'Science_Block':   {'Lecture_Hall_B': 6, 'Cafe': 5,
                        'Engineering_Lab': 7, 'Ring_Road_N': 4},
    'Engineering_Lab': {'Science_Block': 7, 'Workshop': 5, 'Ring_Road_N': 6},
    'Workshop':        {'Engineering_Lab': 5, 'Ring_Road_S': 4},
    'Ring_Road_N':     {'Ring_Road_E': 3, 'Science_Block': 4, 'Engineering_Lab': 6},
    'Ring_Road_E':     {'Ring_Road_N': 3, 'Ring_Road_S': 3, 'Sports_Centre': 5},
    'Ring_Road_S':     {'Ring_Road_E': 3, 'Ring_Road_W': 3,          # LOOP
                        'Workshop': 4, 'South_Lodge': 5},
    'Ring_Road_W':     {'Ring_Road_S': 3, 'Ring_Road_N': 3, 'Sports_Centre': 4},
    'Sports_Centre':   {'Ring_Road_E': 5, 'Ring_Road_W': 4, 'Car_Park': 7},
    'South_Lodge':     {'Main_Gate': 4, 'Ring_Road_S': 5},
    # Maintenance_Road: RESTRICTED — not modelled
}

# Admissible heuristic: straight-line time estimates (minutes) to Engineering_Lab.
# h(n) <= true cost for all n — verified against UCS across all test cases.
# For other goal nodes the heuristic defaults to 0 (A* degrades to UCS, still optimal).
heuristic_to_engineering = {
    'Main_Gate': 18,   'Reception': 16,      'Car_Park': 14,
    'Admin_Block': 14, 'Library': 13,        'Study_Hub': 12,
    'Cafe': 11,        'Lecture_Hall_A': 13, 'Lecture_Hall_B': 10,
    'Science_Block': 7,'Engineering_Lab': 0, 'Workshop': 5,
    'Ring_Road_N': 6,  'Ring_Road_E': 8,     'Ring_Road_S': 7,
    'Ring_Road_W': 8,  'Sports_Centre': 10,  'South_Lodge': 15,
}

print("Map loaded:", len(campus_map), "nodes.")


# ── CELL 2: Algorithms ───────────────────────────────────────

import heapq
from collections import deque


def a_star_search(graph, heuristic, start, goal):
    """
    A* Search: f(n) = g(n) + h(n)
    Optimal and complete with an admissible heuristic.
    Uses a min-heap frontier and lazy deletion of stale entries.
    Returns: (path, total_cost, nodes_expanded)
    """
    frontier = [(heuristic.get(start, 0), 0, start, [start])]  # (f, g, state, path)
    explored = {}       # state -> best g_cost seen
    nodes_expanded = 0

    while frontier:
        f, g, state, path = heapq.heappop(frontier)

        # Lazy deletion: skip if a cheaper path to this state already found
        if state in explored and explored[state] <= g:
            continue
        explored[state] = g
        nodes_expanded += 1

        if state == goal:
            return path, g, nodes_expanded

        for neighbour, cost in graph.get(state, {}).items():
            new_g = g + cost
            new_f = new_g + heuristic.get(neighbour, 0)
            if neighbour not in explored or explored[neighbour] > new_g:
                heapq.heappush(frontier, (new_f, new_g, neighbour, path + [neighbour]))

    return None, float('inf'), nodes_expanded   # no path exists


def uniform_cost_search(graph, start, goal):
    """
    UCS: expands nodes in order of path cost only (no heuristic).
    Also optimal, but expands more nodes than A* when a heuristic is available.
    Returns: (path, total_cost, nodes_expanded)
    """
    frontier = [(0, start, [start])]
    explored = {}
    nodes_expanded = 0

    while frontier:
        g, state, path = heapq.heappop(frontier)
        if state in explored and explored[state] <= g:
            continue
        explored[state] = g
        nodes_expanded += 1
        if state == goal:
            return path, g, nodes_expanded
        for neighbour, cost in graph.get(state, {}).items():
            new_g = g + cost
            if neighbour not in explored or explored[neighbour] > new_g:
                heapq.heappush(frontier, (new_g, neighbour, path + [neighbour]))

    return None, float('inf'), nodes_expanded


def breadth_first_search(graph, start, goal):
    """
    BFS: expands shallowest nodes first, ignoring edge costs entirely.
    Finds the path with fewest hops — NOT the lowest-cost path.
    Included to demonstrate why hop-count optimisation is unsuitable here.
    Returns: (path, hop_count, nodes_expanded)
    """
    if start == goal:
        return [start], 0, 1
    frontier = deque([(start, [start])])
    explored = {start}
    nodes_expanded = 0

    while frontier:
        state, path = frontier.popleft()
        nodes_expanded += 1
        for neighbour in graph.get(state, {}):
            if neighbour not in explored:
                new_path = path + [neighbour]
                if neighbour == goal:
                    return new_path, len(new_path) - 1, nodes_expanded
                explored.add(neighbour)
                frontier.append((neighbour, new_path))

    return None, float('inf'), nodes_expanded


print("Algorithms loaded: A*, UCS, BFS.")


# ── CELL 3: Helper — print a single result ───────────────────

def print_result(algorithm, path, cost, nodes_expanded, graph=None, is_bfs=False):
    """Pretty-print a single algorithm result."""
    print(f"\n  {algorithm}")
    print(f"  {'─' * 50}")
    if path is None:
        print("  No path found.")
        return
    print(f"  Path:           {' -> '.join(path)}")
    if is_bfs:
        # BFS cost = hops; compute actual minute cost separately for comparison
        actual = sum(graph[path[i]][path[i+1]] for i in range(len(path) - 1))
        print(f"  Hops:           {cost}")
        print(f"  Actual cost:    {actual} min  ← BFS ignores costs!")
    else:
        print(f"  Total cost:     {cost} min")
    print(f"  Nodes expanded: {nodes_expanded}")


# ── CELL 4: Test suite ───────────────────────────────────────

def run_tests(graph, heuristic):
    """
    Run all test cases. Each case runs A*, UCS, and BFS, then checks
    whether A* and UCS agree on cost (confirms heuristic admissibility).
    """
    test_cases = [
        # (label,                                          start,            goal)
        ("Standard:  Main Gate -> Engineering Lab",       "Main_Gate",       "Engineering_Lab"),
        ("Standard:  Reception -> Workshop",              "Reception",       "Workshop"),
        ("Loop:      Sports Centre -> Workshop",          "Sports_Centre",   "Workshop"),
        ("One-way:   Library -> Admin Block (detour)",    "Library",         "Admin_Block"),
        ("Edge case: Start equals Goal",                  "Cafe",            "Cafe"),
        ("Edge case: Adjacent nodes",                     "Workshop",        "Engineering_Lab"),
        ("Long:      South Lodge -> Ring Road N",         "South_Lodge",     "Ring_Road_N"),
    ]

    for label, start, goal in test_cases:
        print(f"\n{'=' * 60}")
        print(f"  TEST: {label}")
        print(f"  Start: {start}  |  Goal: {goal}")
        print(f"{'=' * 60}")

        # Edge case: start == goal
        if start == goal:
            print("  [Trivially solved — start equals goal, cost = 0]")
            continue

        # Use the domain heuristic for Engineering_Lab; zero heuristic otherwise
        h = heuristic if goal == "Engineering_Lab" else {n: 0 for n in graph}

        a_path,  a_cost,  a_exp  = a_star_search(graph, h, start, goal)
        u_path,  u_cost,  u_exp  = uniform_cost_search(graph, start, goal)
        b_path,  b_hops,  b_exp  = breadth_first_search(graph, start, goal)

        print_result("A* Search",            a_path, a_cost, a_exp)
        print_result("Uniform Cost Search",  u_path, u_cost, u_exp)
        print_result("Breadth-First Search", b_path, b_hops, b_exp, graph=graph, is_bfs=True)

        # Optimality check: A* and UCS must agree when heuristic is admissible
        if a_path and u_path:
            status = "✓ MATCH" if a_cost == u_cost else f"✗ MISMATCH (A*={a_cost}, UCS={u_cost})"
            print(f"\n  Optimality check (A* vs UCS): {status}")


run_tests(campus_map, heuristic_to_engineering)


# ── CELL 5: Interactive query ────────────────────────────────
# Enter any start and goal node to find a route.

def query(start, goal, graph=campus_map, heuristic=heuristic_to_engineering):
    """
    Run a single route-finding query with all three algorithms.
    Change start and goal to any node names in campus_map.
    """
    print(f"\n{'=' * 60}")
    print(f"  QUERY: {start} -> {goal}")
    print(f"{'=' * 60}")

    if start not in graph:
        print(f"  Error: '{start}' is not a valid node.")
        print(f"  Valid nodes: {sorted(graph.keys())}")
        return
    if goal not in graph:
        print(f"  Error: '{goal}' is not a valid node.")
        print(f"  Valid nodes: {sorted(graph.keys())}")
        return
    if start == goal:
        print("  Start equals goal — cost = 0.")
        return

    h = heuristic if goal == "Engineering_Lab" else {n: 0 for n in graph}

    a_path, a_cost, a_exp = a_star_search(graph, h, start, goal)
    u_path, u_cost, u_exp = uniform_cost_search(graph, start, goal)
    b_path, b_hops, b_exp = breadth_first_search(graph, start, goal)

    print_result("A* Search",            a_path, a_cost, a_exp)
    print_result("Uniform Cost Search",  u_path, u_cost, u_exp)
    print_result("Breadth-First Search", b_path, b_hops, b_exp, graph=graph, is_bfs=True)

    if a_path and u_path:
        status = "✓ MATCH" if a_cost == u_cost else f"✗ MISMATCH (A*={a_cost}, UCS={u_cost})"
        print(f"\n  Optimality check (A* vs UCS): {status}")


# ── Change these two values to query any route ───────────────
START = "Main_Gate"
GOAL  = "Engineering_Lab"
# ─────────────────────────────────────────────────────────────

query(START, GOAL)
```
